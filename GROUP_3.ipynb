```python
clean=food.drop(columns= ['calories_chicken', 'comfort_food', 'comfort_food_reasons', 'calories_chicken', 'calories_scone','cuisine', 'diet_current',
                            'eating_changes', 'eating_changes_coded1', 'father_education', 'father_profession', 'fav_cuisine',
                            'fav_cuisine_coded', 'fries', 'food_childhood', 'healthy_feeling', 'healthy_meal',
                            'ideal_diet', 'indian_food', 'italian_food', 'life_rewarding', 'meals_dinner_friend',
                            'mother_profession', 'mother_education', 'thai_food', 'tortilla_calories', 'turkey_calories', 'waffle_calories',
                            'persian_food','comfort_food_reasons', 'pay_meal_out', 'parents_cook', 'comfort_food_reasons_coded', 'ethnic_food',
                          	'fav_food', 'greek_food', 'soup', 'type_sports', 'sports'])
clean.head()
```

```python
#Checking for duplicates
clean.duplicated().sum()
```

```python
  #Checking for null values
clean.isnull().sum()
```

```python
#Column Renaming
clean.rename(columns={'eating_changes_coded': 'eating_changes'}, inplace=True)
clean.rename(columns={'veggies_day': 'veggies_intake'}, inplace=True)
clean.rename(columns={'calories_day': 'calories_intake'}, inplace=True)
clean.rename(columns={'comfort_food_reasons_coded.1': 'comfort_food_reason'}, inplace=True)
clean.rename(columns={'cook': 'cooking_routine'}, inplace=True)
clean.rename(columns={'drink': 'drink_choice'}, inplace=True)
clean.rename(columns={'fruit_day': 'fruit_intake'}, inplace=True)
clean.rename(columns={'vitamins': 'vitamins_intake'}, inplace=True)
clean.rename(columns={'on_off_campus': 'living_situation'}, inplace=True)
clean.rename(columns={'soup': 'soup_choice'}, inplace=True)
clean.rename(columns={'diet_current_coded': 'health_status'}, inplace=True)
clean.rename(columns={'ideal_diet_coded': 'ideal_diet'}, inplace=True)
```

```python
#Rows renaming
clean['eating_changes'] = clean['eating_changes'].replace({1:'worse', 2:'better', 3:'neutral', 4:'unclear'})
clean['employment'] = clean['employment'].replace({1:'Full Time', 2:'Part Time', 3:'No', 4:'Other'})
clean['exercise'] = clean['exercise'].replace({1:'Everyday',2:'Often', 3:'Weekly', 4:'Sometimes', 5:'Never'})
clean['comfort_food_reason'] = clean['comfort_food_reason'].replace({1: 'stress', 2: 'boredom', 3: 'depression', 4: 'hunger', 5: 'laziness', 6: 'cold weather', 7: 'happiness', 8: 'watching tv', 9: 'none'})
clean['cooking_routine'] = clean['cooking_routine'].replace({1: 'Every day', 2: 'Often', 3: 'Sometimes', 4: 'Once in a while', 5: 'Never'})
clean['income'] = clean['income'].replace({1:'less than $15,000', 2:'$15,001 to $30,000', 3:'$30,001 to $50,000', 4:'$50,001 to $70,000', 5:'$70,001 to $100,000', 6:'higher than $100,000'})
clean['marital_status'] = clean['marital_status'].replace({1:'Single', 2:'In a relationship', 3:'Cohabiting', 4:'Married', 5:'Divorced', 6:'Widowed'})
clean['living_situation'] = clean['living_situation'].replace({1:'On campus', 2:'Out of campus', 3:'Living with parents', 4:'Own a house'})
clean['nutritional_check'] = clean['nutritional_check'].replace({1:'Never', 2:'On certain products only', 3:'Very rarely', 4:'On most products', 5:'Always'})
clean['self_perception_weight'] = clean['self_perception_weight'].replace({6:"I don't use the terms", 5:'Overweight', 4:'Slightly overweight', 3:'Just right', 2:'Very fit', 1:'Slim'})
clean['grade_level'] = clean['grade_level'].replace({1:'Freshman',2:'Sophomore',3:'Junior', 4:'Senior'})
clean['ideal_diet'] = clean['ideal_diet'].replace({1:'portion control', 2:'eating healthier', 3:'balance', 4:'less sugar', 5:'home cooked/organic', 6:'current diet', 7:'more protein', 8:'unclear'})
clean['coffee'] = clean['coffee'].replace({1:'Creamy frappucino', 2:'Coffee'})
clean['breakfast'] = clean['breakfast'].replace({1:'Cereals', 2:'Doughnuts'})
clean['Gender'] = clean['Gender'].replace({1:'Female', 2:'Male'})
clean['calories_intake'] = clean['calories_intake'].replace({1:'No idea', 2:'Not important', 3:'Moderately important', 4:'Very important'})
```

```python
clean.head()
```

```python
#Removing null values in columns
clean['drink_choice'] = clean['drink_choice'].replace({np.nan: ''})
clean['exercise'] = clean['exercise'].replace({np.nan: ''})
clean['living_situation'] = clean['living_situation'].replace({np.nan: ''})
clean['self_perception_weight'] = clean['self_perception_weight'].replace({np.nan: ''})
clean['income'] = clean['income'].replace({np.nan: ''})
clean['calories_intake'] = clean['calories_intake'].replace({np.nan: ''})
clean['marital_status'] = clean['marital_status'].replace({np.nan: ''})
clean['employment'] = clean['employment'].replace({np.nan: ''})
clean['cooking_routine'] = clean['cooking_routine'].replace({np.nan: ''})
clean['GPA'] = clean['GPA'].replace({'3.79 bitch': '3.79', np.nan: '', 'Personal': '', 'Unknown': ''})
clean['weight'] = clean['weight'].replace({np.nan: '', 'Not sure, 240': 240, "I'm not answering this. ": '', '144 lbs': 144})
```

```python
#Checking for null values
clean.isnull().sum()
```

```python
#Changing the data type and filling null values for weight
clean['weight']= pd.to_numeric(clean['weight'], errors ='coerce')
mean_weight = clean['weight'].mean()
clean['weight'].fillna(mean_weight, inplace = True)
```

```python
#Changing the data type and filling null values for weight
clean['drink_choice']= pd.to_numeric(clean['drink_choice'], errors ='coerce')
mean_weight = clean['drink_choice'].mean()
clean['drink_choice'].fillna(mean_weight, inplace = True)
```

# **Visualizations**

> The visualizations of the datasets were categorised into three:


*   ***Univariate: Single column analysis***
*   ***Bivariate: Double column analysis***
*   ***Multivariate: Multiple column analysis.***





**Univariate**

```python
#Plot showing how the percentage of students at different levels
clean["grade_level"].value_counts().plot(kind='pie', colors=['azure', 'lime','darkgreen','teal'], autopct='%1.2f%%', startangle=90)

# Adding a title
plt.title('Grade_level Distribution', weight= 'bold')

# Display the plot
plt.show()
```

```python
#Plot showing the distribution of the student's GPA.
sns.histplot(clean['GPA'], bins=20, kde=True, color='teal')
plt.title("Distribution of GPA", weight= 'bold')
plt.xlabel("GPA")
plt.show()
```

***From the figure above, it is depicted that the distribution of the student's GPA is LEFT SKEWED/NEGATIVE SKEWED because the tail of the distribution is on the left and most of the data is on the right.***

```python
#Plot showing how the cooking routine of students affects their GPA.
ax= sns.barplot(data=clean, x='cooking_routine', y='GPA', hue= 'GPA', legend= True, palette='viridis')
plt.xlabel('Cooking routine')
plt.ylabel('GPA')
plt.xticks(rotation=15)
plt.title("Impact of cooking on \nStudent's GPA", weight='bold')
plt.show()
```

***From the figure above, it is observed that the cooking routine of the students has little or no effect on their GPA.***

```python
#Plot showing how the impact of daily calories intake on the weight of students
sns.histplot(data=clean, x='weight', hue='calories_intake', kde=True, multiple='stack', palette='coolwarm')
plt.xlabel('Weight')
plt.ylabel('Calories Intake')
plt.title("Impact of Daily Calories Intake\n on Student's Weight", weight='bold')
plt.show()
```

***From the figure above, those who claim daily calories intake are "very important/moderately important" weigh between 100-250; while those who claim daily calories intake are "not important" weigh between 100-200.***

***It is also depicted that the distribution of the student's GPA is RIGHT SKEWED/POSITIVE SKEWED because the tail of the distribution is on the right and most of the data is on the left.***

```python
#Plot showing how the student's self-perceived weight relates to their actual weight
ax=sns.barplot(data=clean, y='weight', x='self_perception_weight', hue= 'weight');
plt.title("Correlation between the Weight of Students \nand their Self-Perceived Weight", weight='bold')
plt.xticks(rotation=45)
plt.show()
```

***From the figure above, those who claim they are "just right/slim" weigh below 200; those who claim they are "overweight/slightly overweight" weigh between 200-270; those who claim they are "very fit" weigh between 110-200;.***

```python
#Plot showing how student's income affects their health's status
pd.crosstab(clean['health_status'], clean['income']).plot(kind='barh', stacked=True, cmap= 'coolwarm')
plt.xlabel('Income')
plt.ylabel('Health Status')
plt.title('Impact of Income on Health', weight='bold')
plt.show()
```

```python
#Plotting the categorical columns to show the effects of healthy living on student's peformance
features = ["exercise","breakfast","eating_changes","comfort_food_reason","living_situation", "calories_intake", "nutritional_check"]
fig,axes = plt.subplots(1,7,figsize=(20,8),tight_layout=False)
for feature,ax in zip(features, axes):
  plt.sca(ax)
  sns.boxplot(x=clean[feature], y=clean["GPA"], color = 'teal')
  plt.xlabel(feature, fontsize= 15, color = 'black')
  plt.xticks(fontsize = 9, rotation=45)
plt.suptitle("Effect of Healthy Living on Student's GPA", fontsize= 20, weight='bold')
plt.show()
```

```python
#Plotting the numerical columns to show the effects of healthy eating habits
features1 = ["eating_out","fruit_intake","veggies_intake","vitamins_intake","health_status", "drink_choice","weight"]
fig,axes=plt.subplots(1,7,figsize=(15,4),tight_layout=True)
for feature1,ax in zip(features1, axes):
  plt.sca(ax)
  sns.kdeplot(clean.loc[clean["GPA"] <=3.00, feature1], label="<=3",fill=True)
  sns.kdeplot(clean.loc[clean["GPA"] <=4.00, feature1], label="<=4",fill=True)
  plt.legend(title="GPA")
plt.suptitle("Effect of Healthy Eating on Student's GPA", fontsize=16, weight='bold')
plt.show()
```

***Eating out*** : 1-Never,  2-Rarely,  3-Sometimes,  4-Occasionally,   5-Everyday

***Veggies_intake*** : 5-Very likely,  4-Likely,  3-Neutral,  2-Unlikely,
  1-Very unlikely

***Fruit_intake***: 5-Very likely,  4-Likely,  3-Neutral,  2-Likely,  1-Very unlikely

 ***Vitamins_intake***:1-Yes,  2-No

***Drink_choice*** : 1-Orange juice,  2-Soda

 ***Health_status*** : 4-Not sure,  3-Moderately,  2-Unhealthy,  1-Healthy
